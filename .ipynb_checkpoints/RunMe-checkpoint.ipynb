```python
try:
    import cv2
except ImportError:
    print "OpenCV NOT Found! Please see the instructions above to install (pip install cv2 or similar variations won't work.)"
import tools as t
import classifier as c
import display_features as d
import matplotlib.pyplot as plt

e = t.environment("images.csv") # This "e" variable will hold our environment which we'll later reuse.
```



Now we'll extract features from 3 sample images in our dataset by running the code below. 



```python
%matplotlib inline
img = d.get_features("apple-tree-429213_640.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.figure(2)
img2 = d.get_features("goose-15032_640.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.figure(3)
img3 = d.get_features("cat-1373895_640.jpg")
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
plt.imshow(img3)
```


Our dataset consists of 100 images of apples, 100 images of birds, and 100 images of cats.  Let's see how many features we can extract from every photo in our exploratory visualization below.



```python

allimgs = e.imgs['apple'] + e.imgs['bird'] + e.imgs['cat']
features = []
detector = cv2.FeatureDetector_create("SIFT")
extractor = cv2.DescriptorExtractor_create("SIFT")
for i in allimgs:
    kps = detector.detect(i.data)
    (kps, descs) = extractor.compute(i.data, kps)
    features.append(len(descs))
plt.hist(features, 25)
plt.xlabel("# of Features")
plt.ylabel("# of images")
plt.title("SIFT Features extracted from ALL Photos")
plt.axis([0, 7000, 0, 70])
```


Now it's time to test some parameter combinations.  Only a small fraction of parameter combinations that were tested will be shown in the subsequent sections for demonstration.  

The first parameter, trainsize = 50, will take 50 random images from each category for use in the Training Set.  The 2nd parameter, testsize = 50, will take 50 random images not selected for the Training Set to be put into the Testing Set.  The 3rd parameter, sets = 10, will do 10 different training/testing splits and average the results from each split.

Note: "Average Training/Testing Time" is the total amount of time required to train/test the entire set, not time per picture!

Depending on your computer speed, this may take a couple of minutes.

```python
from sklearn.naive_bayes import BernoulliNB
e.benchmark(trainsize=50, testsize=50, sets=10, clf=BernoulliNB(), feature_finder="ORB", convert_grey=False, 
            rootsift=False, pca_before_kmeans=True, kclusters=100, pca_ratio=0.25, tfidf=True, incremental_threshold=7)
```


Now let's try changing the Classifier from BernoulliNB (Bernoulli Naive-Bayes) to SGDClassifier (Stochastic Gradient Descent):


```python
from sklearn.linear_model import SGDClassifier
e.benchmark(trainsize=50, testsize=50, sets=10, clf=SGDClassifier(), feature_finder="ORB", convert_grey=False, 
            rootsift=False, pca_before_kmeans=True, kclusters=100, pca_ratio=0.25, tfidf=True, incremental_threshold=7)
```


The following parameters are close to the final parameters chosen for the model. (The only difference is that, in the interest of time, feature_finder="ORB" instead of feature_finder="SIFT" was chosen here.)


```python
e.benchmark(trainsize=50, testsize=50, sets=10, clf=SGDClassifier(), feature_finder="ORB", convert_grey=True, 
            rootsift=True, pca_before_kmeans=True, kclusters=200, pca_ratio=0.5, tfidf=False, incremental_threshold=25)
```


So far, we've only worked with a single dataset.  Let's try opening a different set of images: http://www.vision.caltech.edu/Image_Datasets/Caltech101

Note the images have already been downloaded into the package; the link above is included for citation and further reference.

I've created a child class to environment: caltech_environment, which makes it more convenient for loading images and automatically assigning them their classifications by the name of their folders.  We'll look at these categories:

- airplanes
- brain
- crab

And take 20 random images from each in the training set and another 20 random images from each in the testing set for a total of 10 different runs.

```python
ec = t.caltech_environment(dirlist=["airplanes","brain","crab"])
ec.benchmark(trainsize=20, testsize=20, sets=10, clf=SGDClassifier(), feature_finder="ORB", convert_grey=True, 
            rootsift=True, pca_before_kmeans=True, kclusters=200, pca_ratio=0.5, tfidf=False, incremental_threshold=25)
```


Now let's try adding a 4th category: pizza.  This one has very dense and varied features which is expected to drastically reduce the F1 score.  Let's see...


```python
ec2 = t.caltech_environment(dirlist=["airplanes","brain","crab","pizza"])
ec2.benchmark(trainsize=20, testsize=20, sets=10, clf=SGDClassifier(), feature_finder="ORB", convert_grey=True, 
            rootsift=True, pca_before_kmeans=True, kclusters=200, pca_ratio=0.5, tfidf=False, incremental_threshold=25)

```


This time, let's swap "pizza" for "stop_sign" which has far fewer features:


```python
ec3 = t.caltech_environment(dirlist=["airplanes","brain","crab","stop_sign"])
ec3.benchmark(trainsize=20, testsize=20, sets=10, clf=SGDClassifier(), feature_finder="ORB", convert_grey=True, 
            rootsift=True, pca_before_kmeans=True, kclusters=200, pca_ratio=0.5, tfidf=False, incremental_threshold=25)
```

The F1 score here is probably higher than the F1 score in the previous example although not as high as the one with only 3 classes.  This may be due to the fact that a classifier that just randomly guesses the class will achieve an F1 score of 1/N, where N is the # of classes.  Hence, the F1 score will be biased higher with a smaller # of classes.  Let's see if we can achieve an even higher F1 Score with only 2 classes below:

```python
ec4 = t.caltech_environment(dirlist=["airplanes","brain"])
ec4.benchmark(trainsize=20, testsize=20, sets=10, clf=SGDClassifier(), feature_finder="ORB", convert_grey=True, 
            rootsift=True, pca_before_kmeans=True, kclusters=200, pca_ratio=0.5, tfidf=False, incremental_threshold=25)
```


Next, let's take a look at the average # of SIFT features in every one of the classes of images we've looked at so far.


```python
print "Apples: {}".format(d.avg_features(e.imgs['apple']))
print "Birds: {}".format(d.avg_features(e.imgs['bird']))
print "Cats: {}".format(d.avg_features(e.imgs['cat']))
print "Airplanes: {}".format(d.avg_features(ec.imgs['airplanes']))
print "Brains: {}".format(d.avg_features(ec.imgs['brain']))
print "Crabs: {}".format(d.avg_features(ec.imgs['crab']))
print "Pizzas: {}".format(d.avg_features(ec2.imgs['pizza']))
print "Stop Signs: {}".format(d.avg_features(ec3.imgs['stop_sign']))
```


Now let's actually try identifying some images.  Let's start with the Airplanes, Brains, and Crabs set.  We'll train using the maximum # of images in the training set and try to predict 3 images I've found off of the Internet below.


```python

cc = ec.train_all(clf=SGDClassifier(), feature_finder="SIFT", convert_grey=True, 
            rootsift=True, pca_before_kmeans=True, kclusters=200, pca_ratio=0.5, tfidf=False, incremental_threshold=25)

cc.predict_and_visualize("newimages/airplane1.jpg","airplane1.jpg")
cc.predict_and_visualize("newimages/airplane2.jpg","airplane2.jpg")
cc.predict_and_visualize("newimages/brain1.png","brain1.png")
cc.predict_and_visualize("newimages/brain2.png","brain2.png")
cc.predict_and_visualize("newimages/crab1.jpg","crab1.jpg")
cc.predict_and_visualize("newimages/crab2.jpg","crab2.jpg")


plt.imshow(cv2.cvtColor(cv2.imread("airplane1.jpg"), cv2.COLOR_BGR2RGB))
plt.figure(5)
plt.imshow(cv2.cvtColor(cv2.imread("airplane2.jpg"), cv2.COLOR_BGR2RGB))
plt.figure(6)
plt.imshow(cv2.cvtColor(cv2.imread("brain1.png"), cv2.COLOR_BGR2RGB))
plt.figure(7)
plt.imshow(cv2.cvtColor(cv2.imread("brain2.png"), cv2.COLOR_BGR2RGB))
plt.figure(8)
plt.imshow(cv2.cvtColor(cv2.imread("crab1.jpg"), cv2.COLOR_BGR2RGB))
plt.figure(9)
plt.imshow(cv2.cvtColor(cv2.imread("crab2.jpg"), cv2.COLOR_BGR2RGB))
plt.figure(10)
```
